```python
#pip install pandas
#para tranajar con archivos excel, CSV, etc
import pandas as pd
#excel llevado a python
df = pd.read_csv('df_total.csv', encoding = 'UTF-8')
df
```

```python
#cogemos la noticia 3
df["news"][3]
```

```python
# lo primero es Separamos los datos en variables de entrada y etiquetas
X = df['news']
y = df['Type']
```

```python
#vemos cuantos hay
print(df['Type'].value_counts())
```

```python
#Esta librería permite dividir un conjunto de datos para entrenamiento y otro para probar el modelo
from sklearn.model_selection import train_test_split
```

```python
#entrenamos el modelo con el 80% de los datos y lo probamos con el 20%
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
#Vamos a vectorizar los datos, con CountVectorizer contamos cuantas veces están las palabras
from sklearn.feature_extraction.text import CountVectorizer
```

```python
vectorizer = CountVectorizer()

#Xtransformados o vectorizados
X_train_trasformed = vectorizer.fit_transform(X_train)
X_test_trasformed = vectorizer.transform(X_test)
```

```python
X_train_trasformed
```

```python
#Ejemplo visual de trabajo vectorizado
X_train_trasformed_dense = X_train_trasformed.toarray()
print(X_train_trasformed_dense)
```

```python
from sklearn.naive_bayes import MultinomialNB
from sklearn import metrics

# Vamos a meter los X de entrenamiento a un modelo multinomial para entrenarlo y crear un modelo 
# de predicción en base a los vectores cual va a ser el tipo de noticia
model = MultinomialNB()
#model.fit(Variable independiente, Varible dependiente)
model.fit(X_train_trasformed, y_train)
#y:pred para medir el rendimeinto de modelo
y_pred = model.predict(X_test_trasformed)
print(metrics.accuracy_score(y_test, y_pred))
```

```python
##################Stemming#########################

import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import SnowballStemmer

nltk.download('punkt')
nltk.download('stopwords')
```

```python
stemmer = SnowballStemmer('spanish')
def tokenize_and_stem(text):
    #aplicar un tokenize, es decir separar el texto en palabras
    tokens = word_tokenize(text.lower())
    stems = [stemmer.stem(token) for token in tokens if token.isalpha()]
    return ' '.join(stems)
```

```python
df['news_stemmer'] = df['news'].apply(tokenize_and_stem)
```

```python
df['news_stemmer']
```

```python
print(df['news_stemmer'][3])
```

```python
#Ahora hacemos lo mismo pero con los datos stimeados

X = df['news_stemmer']
y = df['Type']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
vectorizer = CountVectorizer()
X_train_trasformed = vectorizer.fit_transform(X_train)
X_test_trasformed = vectorizer.transform(X_test)

model = MultinomialNB()
model.fit(X_train_trasformed, y_train)

#medimos rendimiento
y_pred = model.predict(X_test_trasformed)
print(metrics.accuracy_score(y_test, y_pred))
```

```python
##################Lemmatization#########################
import spacy
nlp = spacy.load('es_core_news_sm')
```

```python
def lemmatize_text(text):
    doc = nlp(text.lower())
    lemmas = [token.lemma_ for token in doc if token.is_alpha]
    return ' '.join(lemmas)
```

```python
df['news_lemma'] = df['news'].apply(lemmatize_text)
```

```python
df['news_lemma']
```

```python
df['news_lemma'][3]
```

```python
#Ahora hacemos lo mismo pero con los datos de lemma

X = df['news_lemma']
y = df['Type']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
vectorizer = CountVectorizer()
X_train_trasformed = vectorizer.fit_transform(X_train)
X_test_trasformed = vectorizer.transform(X_test)

model = MultinomialNB()
model.fit(X_train_trasformed, y_train)

#medimos rendimiento
y_pred = model.predict(X_test_trasformed)
print(metrics.accuracy_score(y_test, y_pred))
```
