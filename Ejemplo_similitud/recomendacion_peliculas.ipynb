```python
#Similitud de vectores - Recomendador de Películas
import pandas as pd

#Si el archivo está separado por comas -> df = pd.read_csv('movie_metadata.csv')
#Para nuestro caso
df = pd.read_csv('movie_metadata.csv', sep=';', encoding='utf-8-sig')
# print(df.head())
df
```

```python
#vamos a cambiar los caracteres "|" verticales
df['genero'] = df['genero'].str.replace('|', ' ')
df['plot_keywords'] = df['plot_keywords'].str.replace('|', ' ')
df
```

```python
df['texto'] = df[['genero', 'plot_keywords']].apply(lambda row: ' '.join(row.values.astype(str)), axis=1)
df
```

```python
row = df[['genero', 'plot_keywords', 'texto']].iloc[0]
row
```

```python
print(df['texto'].iloc[0])
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity, euclidean_distances
#Cantidad máxima de toquen 2000, las peliculas no deben tener mas de eso
```

```python
tfidf = TfidfVectorizer(max_features=2000)
X = tfidf.fit_transform(df['texto'])
X
```

```python
peliculas = pd.Series(df.index, index=df['movie_title'])
#elimina estapcios de principio y fin
peliculas.index = peliculas.index.str.strip()
peliculas
```

```python
indice = peliculas['The Dark Knight Rises']
indice2 = peliculas['Avatar']
print(indice, indice2)
```

```python
consulta = X[indice]
consulta
```

```python
print(consulta)
```

```python
consulta.toarray()
```

```python
similitud = cosine_similarity(consulta, X)
similitud
```

```python
similitud[0][3]
```

```python
similitud = similitud.flatten()
similitud[3]
```

```python
#pip install matplotlib
import matplotlib.pyplot as plt

#graficar similitud
plt.plot(similitud)
```
